```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data.csv')
pd.set_option("display.max_columns",25)
print('Shape of dataframe',df.shape)
df
```

```python
# Remove duplicates
df = df.drop_duplicates(keep='first')
print('Shape of data frame:', df.shape)
df
```

```python
# Check for null values
df.info()

# There are null values in 'Car','Bar','CoffeeHouse','CarryAway','RestaurantLessThan20','Restaurant20To50'
```

```python
# Null Value Dataframe

missing_value = df.isnull().sum()*100/len(df)
missing_df = pd.DataFrame({'MissingValueCount': df.isnull().sum(),'MissingPercentage': missing_value})
missing_df[missing_df['MissingValueCount']!=0]
```

```python
# Handling Missing Values with Mode Imputation

# There is more than 99% missing values in 'car' rows, so drop this column
df1 = df.drop('car', axis = 1)

df1['Bar'] = df1['Bar'].fillna(df1['Bar'].value_counts().index[0])
df1['CoffeeHouse'] = df1['CoffeeHouse'].fillna(df1['CoffeeHouse'].value_counts().index[0])
df1['CarryAway'] = df1['CarryAway'].fillna(df1['CarryAway'].value_counts().index[0])
df1['RestaurantLessThan20'] = df1['RestaurantLessThan20'].fillna(df1['RestaurantLessThan20'].value_counts().index[0])
df1['Restaurant20To50'] = df1['Restaurant20To50'].fillna(df1['Restaurant20To50'].value_counts().index[0])
```

```python
data = df1.corr()
data

# toCoupon_GEQ5min shows no correlation to any other variables. Thus, we need to drop this column
# direction_same has exactly the same correlation with direction_opp. Thus, drop one of them
```

```python
df1 = df1.drop(['toCoupon_GEQ5min', 'direction_opp'], axis = 1)
```

# EDA

```python
# Perform feature engineering for 'toCoupon'
# 0 -> use coupon with driving distance of 15 mins
# 1 -> use coupon with driving distance of 25 mins
# 2 -> use coupon with driving distance of both 15 and 25 mins
# 3 -> don't use both coupon 

tocoupon = []
for index, row in df1.iterrows():
    if (row['toCoupon_GEQ15min']==1 and row['toCoupon_GEQ25min']==0):
        tocoupon.append(0)
    elif (row['toCoupon_GEQ15min']==0 and row['toCoupon_GEQ25min']==1):
        tocoupon.append(1)
    elif (row['toCoupon_GEQ15min']==1 and row['toCoupon_GEQ25min']==1):
        tocoupon.append(2)
    else:
        tocoupon.append(3)
df1['tocoupon'] = tocoupon
df1
```

```python
df1 = df1.drop(['toCoupon_GEQ15min','toCoupon_GEQ25min'], axis = 1)
df1
```

```python
# Count the percentage of the Target variables

accept = df1['Accept(Y/N?)'].value_counts()[1]
accept_pct = accept/len(df1)
reject = df1['Accept(Y/N?)'].value_counts()[0]
reject_pct = reject/len(df1)
print('Accepted Coupon: ', accept, ' or ', accept_pct*100, '%')
print('Rejected Coupon: ', reject, ' or ', reject_pct*100, '%')
```

### Univariate Analysis

```python
# Univariate Analysis on Destination

df_desti = pd.DataFrame(sorted(list(df1['destination'].unique())), columns = ['destination'])
df_desti['totalcount'] = list(df1.groupby('destination')['Accept(Y/N?)'].count())
df_desti['totalpct'] = round(df_desti['totalcount']/df1['destination'].shape[0]*100, 3)
df_desti['accepted'] = list(df1[df1['Accept(Y/N?)']==1].groupby('destination')['Accept(Y/N?)'].count())
df_desti['rejected'] = list(df1[df1['Accept(Y/N?)']==0].groupby('destination')['Accept(Y/N?)'].count())
df_desti['acceptedpct'] = round(df_desti['accepted'] / df_desti['totalcount'] * 100, 3)
df_desti['rejectedpct'] = round(df_desti['rejected'] / df_desti['totalcount'] * 100, 3)

f, ax = plt.subplots(figsize=(8, 6))

# Sort the DataFrame by totalcount in descending order
data = df_desti.sort_values("totalcount", ascending=False)

# Plot the total count
sns.set_color_codes("pastel")
sns.barplot(x="totalcount", y="destination", data=data,
            label="TotalCount", color = 'b')

# Plot the Accepted ones
sns.set_color_codes("muted")
sns.barplot(x="accepted", y="destination", data=df_desti,
            label="Accepted", color="b")

# Add a legend and informative axis labels
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlabel="Count", ylabel="Destionation",
       title="Destination - wise Accepted Coupons")
sns.despine(left=True, bottom=True)

plt.show()
df_desti
```

Findings:
- Coupons are mostly offered to people who are not in a hurry or have no urgent destinations, with the highest acceptance rate of 73%
- People who are heading home do not prefer to be offered coupons as evidenced by highest rejection ratio
